
**Solving the SVM problem by inspection**

By inspection we can see that the boundary decision line is the function x2=x1−3. Using the formula wTx+b=0 we can obtain a first guess of the parameters as

w=[1,−1]  b=−3

Using these values we would obtain the following width between the support vectors: 2/√2=√2.

Again by inspection we see that the width between the support vectors is in fact of length 4√2 meaning that these values are incorrect.

Recall that scaling the boundary by a factor of c does not change the boundary line, hence we can generalize the equation as

cx1−x2−3c=0
w=[c,−c]  b=−3c

Plugging back into the equation for the width we get

2/||w|| = 2√2

2/(2√2c) = 4√2

c=1/4

Hence the parameters are in fact

w=[1/4,−1/4]  b=−3/4


SKlearn 解法

```python
from sklearn.svm import SVC

X = np.array([[3,4],[1,4],[2,3],[6,-1],[7,-1],[5,-3]] )
y = np.array([-1,-1, -1, 1, 1 , 1 ])

clf = SVC(C = 1e5, kernel = 'linear')
clf.fit(X, y) 

print('w = ',clf.coef_)
print('b = ',clf.intercept_)
print('Indices of support vectors = ', clf.support_)
print('Support vectors = ', clf.support_vectors_)
print('Number of support vectors for each class = ', clf.n_support_)
print('Coefficients of the support vector in the decision function = ', np.abs(clf.dual_coef_))
```
